```python
from plotnine import *

import pandas as pd
import math
```

```python
data_file = "RRab_extracted.csv"
data = pd.read_csv(data_file)

data.info()
```

```python
ggplot(data, aes("period", "phi31")) +\
    geom_point() +\
    ggtitle("Official OGLE IV phi31")
```

```python
ggplot(data, aes("period", "phi_31")) +\
    geom_point() +\
    ggtitle("Calculated phi31")
```

```python
ggplot(data, aes("R31", "phi31")) +\
    geom_point() +\
    ggtitle("Official OGLE IV phi31 & R31") +\
    xlim(-1,1) +\
    ylim(0, 2 * math.pi)
```

```python
ggplot(data, aes("R_31", "phi_31")) +\
    geom_point() +\
    ggtitle("Calculated phi31 & R31") +\
    xlim(-1,1) +\
    ylim(0, 2 * math.pi)
```

```python
ggplot(data, aes("R21", "phi21")) +\
    geom_point() +\
    ggtitle("Official OGLE IV phi21 & R21") +\
    xlim(-1,1) +\
    ylim(0,6)
```
