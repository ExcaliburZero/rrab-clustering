```python
from plotnine import *

import pandas as pd
```

```python
data_file = "~/Documents/ogle/ogle4/smc/RRab_extra_extracted.csv"
data = pd.read_csv(data_file)

print(data.info())
data.describe()
```

```python
import math

ggplot(data, aes("log_p", "phi31_i_sine")) +\
    geom_point()
```

```python
ggplot(data, aes("log_p", "phi31_v_skow")) +\
    geom_point() +\
    ggtitle("OGLE IV SMC - Calculated V band Phi31")
```

```python
ggplot(data.dropna()) +\
    geom_density(aes("phi31_v_deb"), color="blue") +\
    geom_density(aes("phi31_v_skow"), color="green")
```

```python
ggplot(data, aes("log_p", "metalicity_jk_v")) +\
    geom_point()
```

```python
ggplot(data.dropna()) +\
    geom_density(aes(x = "metalicity_jk_v"), color="red") +\
    geom_density(aes(x = "metalicity_jkzw_v"), color="orange") +\
    geom_density(aes(x = "metalicity_sand"), color="blue") +\
    geom_density(aes(x = "metalicity_neme"), color="purple") +\
    geom_density(aes(x = "metalicity_smol"), color="green") +\
    xlab("[Fe/H]") +\
    ggtitle("OGLE IV SMC Metalicity Formula Distributions")
```

```python
data["diff_jk_neme"] = data.as_matrix(["metalicity_jk_v"]) - data.as_matrix(["metalicity_neme"])
```

```python
ggplot(data, aes(x = "metalicity_neme", y = "diff_jk_neme")) +\
    geom_point() +\
    xlim(-3, -0.5) +\
    ylim(-0.7, 0.7)
```

```python
ggplot(data, aes("metalicity_jkzw_v", "metalicity_sand")) +\
    geom_point() +\
    xlim(-4,1.0) +\
    ylim(-4,1.0)
```
